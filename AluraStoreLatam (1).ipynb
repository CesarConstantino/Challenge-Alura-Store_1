#1. Análisis de facturación



```python
# Crear una función para calcular la facturación
def calcular_facturacion(df):
    return (df['Precio'] + df['Costo de envío']).sum()


# Calcular la facturación total por tienda
facturacion_t1 = calcular_facturacion(tienda)
facturacion_t2 = calcular_facturacion(tienda2)
facturacion_t3 = calcular_facturacion(tienda3)
facturacion_t4 = calcular_facturacion(tienda4)

# Mostrar los resultados
print(f"Facturación Tienda 1: ${facturacion_t1:,.0f}")
print(f"Facturación Tienda 2: ${facturacion_t2:,.0f}")
print(f"Facturación Tienda 3: ${facturacion_t3:,.0f}")
print(f"Facturación Tienda 4: ${facturacion_t4:,.0f}")
```

# 3. Calificación promedio de la tienda


```python
# Promedio de calificación por tienda
promedio_calificacion_t1 = tienda1['Calificación'].mean()
promedio_calificacion_t2 = tienda2['Calificación'].mean()
promedio_calificacion_t3 = tienda3['Calificación'].mean()
promedio_calificacion_t4 = tienda4['Calificación'].mean()

# Mostrar resultados
print("Promedio de calificación de los clientes:")
print(f"Tienda 1: {promedio_calificacion_t1:.2f}")
print(f"Tienda 2: {promedio_calificacion_t2:.2f}")
print(f"Tienda 3: {promedio_calificacion_t3:.2f}")
print(f"Tienda 4: {promedio_calificacion_t4:.2f}")
```

```python
# Calcular los promedios nuevamente
promedios = [
    tienda1['Calificación'].mean(),
    tienda2['Calificación'].mean(),
    tienda3['Calificación'].mean(),
    tienda4['Calificación'].mean()
]

labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colors = sns.color_palette('pastel')

# Gráfico de pastel
plt.figure(figsize=(7, 7))
plt.pie(promedios, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribución de Promedios de Calificación por Tienda')
plt.axis('equal')
plt.show()
```

```python
# Convertimos los promedios en porcentajes del total
total = sum(promedios)
porcentajes = [p / total * 100 for p in promedios]

# Crear DataFrame
calificaciones_pct_df = pd.DataFrame({
    'Tienda': labels,
    'Porcentaje del Promedio': porcentajes
})

# Gráfico de barras horizontales
plt.figure(figsize=(8, 5))
sns.barplot(x='Porcentaje del Promedio', y='Tienda', data=calificaciones_pct_df, palette='mako')
plt.title('Porcentaje del Promedio de Calificación por Tienda')
plt.xlabel('Porcentaje (%)')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

```

# **Interpretación sobre la satisfacción del cliente (calificación promedio):**

Las tiendas muestran niveles de satisfacción relativamente similares, sin embargo, la **Tienda 1** destaca con la calificación promedio más alta, lo que indica que los clientes tienen una mejor percepción de su experiencia de compra allí.
